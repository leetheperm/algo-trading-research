A lot of exhcnages have index funds for large stock exchanges or ETFS, but I haven't seen one that includes cryptocurrencies. 

### Hypothesis

It is widely known in trading/day trading communities that trading anything can be risky. What is not so obvious is that most traders actually LOSE money when comparfed to the wider market. This is true for cryptocurrency trading, stock markets, commodities, basically everything. In 1929 the Wellington Management Company invented the first balanced mutual fund. The theory behind this was that diverification of investments was a less risk averse stratgey and helped simplify the markets for those looking to join. Jack Bogle then built "The Vanguard experiment" on-top of this strategy and updated the model to allow the mutual funds to be 'operated at cost and independently, with their own directors, officers, and staff.' 

As far as I'm aware, there are Crypto ETFs that invest in the top blockchain companies, but less research is done into actual token diversification. There is a bitwise crypto index fund that takes the top 10 cryptocurrencies, but there is a big caveat. It is 'screened for certain risks' it is the weighted by market cap and it rebalances that weight on a monthly basis. It is ironically an awful performer as well when compared to any top 10 cyptocurrency.

My hypothesis will simply take the mean price of the top 50 and track that overtime, compared to the top 50 cryptos at this current time.


### Idea

1. Take currecnt top 50 coins by market cap
2. Find mean price of 50 (all coins prices combines / 50)
3. Track performance of index fund
4. compare with non-aggreagted crypto currencies

```python
import json
import pandas as pd
```

```python
with open("../data_folder/index_data.json", "r",encoding="utf-8") as read_file:
    data = json.load(read_file)

    
name = data['data'][0]['name']
price_btc = data['data'][0]['quote']['USD']
headers = ['price', 'volume 24hr', 'change 1', 'change 24', 'change 7d', 'change 30d', 'change 60d', 'change 90d', 'market cap', 'updated']

# Change this value to sort by different columns
sort_column = "market_cap"
next_crypto = []
name_of_coin = []
# This value can be changed if you would like to see more coins
number_of_iter = 50
for i in range(number_of_iter):
    name = data['data'][i]['name']
    price_crypto = data['data'][i]['quote']['USD']
#     print(name)
#     print(price_crypto)
    next_crypto.append(price_crypto)
    name_of_coin.append(name)
df = pd.DataFrame(next_crypto,index=name_of_coin).sort_values(by=sort_column, ascending=False)
df.round(1)
df.drop(columns=['volume_24h','last_updated'],inplace=True)
df.rename(columns={"percent_change_1h": "change 1h", "percent_change_24h": "change 24h","percent_change_7d": "change 7d" })    
```

```python
mean_price = df["price"].mean().round(2)
mean_1hr = df["percent_change_1h"].mean().round(2)
mean_24hr = df["percent_change_24h"].mean().round(2)
mean_30d = df["percent_change_30d"].mean().round(2)
mean_60d = df["percent_change_60d"].mean().round(2)
mean_90d = df["percent_change_90d"].mean().round(2)


print(f"the mean price of Index50 in {mean_price}")
print(f"the mean 1 hour change of Index50 is {mean_1hr} %")
print(f"the mean 24 hour change of Index50 is {mean_24hr} %")
print(f"the mean 30 day change of Index50 is {mean_30d} %")
print(f"the mean 60 day change of Index50 is {mean_60d} %")
print(f"the mean 90 day change of Index50 is {mean_90d} %")

print("of all crypto's in the last 90 days Index50 was 14th top performer")
```
